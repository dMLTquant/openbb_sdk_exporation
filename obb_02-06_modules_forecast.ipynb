# 02.06 Modules: Forecast

> https://docs.openbb.co/sdk/guides/intros/forecast

```python
from openbb_terminal.sdk import openbb
import pandas as pd
```

```python
EURUSD = openbb.forex.load(
    from_symbol='EUR',
    to_symbol='USD',
    start_date = '2022-01-01',
    interval = '1day')

EURUSD.head(3)
```

```python
EURUSD.head(5)
```

```python
EURUSD.tail(1)
```

```python
openbb.forecast.plot(data=EURUSD, columns = ['Adj Close'])
```

## Theta

```python
help(openbb.forecast.theta)
```

```python
openbb.forecast.theta_chart(data = EURUSD, target_column = 'Adj Close', n_predict= 10)
```

```python
openbb.forecast.regr_chart(EURUSD, target_column = 'Close')
```
