## Dataframe

```python
openbb.ta.obv(data = openbb.stocks.load('QQQ'))
```

## Chart

```python
openbb.ta.obv_chart(data= openbb.stocks.load('QQQ', start_date = '2022-11-18', interval = 5, prepost = True))
```

## Techincal analysis indicators

To use the same data for multiple functions, it is more efficient to first load to a Pandas DataFrame

```python
ticker = 'ES=F'
start = '2020-01-01'

data_df: pd.DataFrame = openbb.stocks.load(f"{ticker}", start_date = f"{start}")
```

### MA (Moving Averages)

```python
openbb.ta.ma_chart( data = data_df['Adj Close'], symbol = 'E-Mini S&P Futures', ma_type = 'SMA', window = [21, 150])
```

### Aroon

```python
openbb.ta.aroon_chart(data_df)
```

### Donchian

```python
openbb.ta.donchian_chart(data_df)
```
