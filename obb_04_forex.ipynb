```python
from openbb_terminal.sdk import openbb
import pandas as pd
#%matplotlib inline
```

```python
currency_pair = openbb.forex.load(
    from_symbol='USD',
    to_symbol='EUR',
    start_date = '2022-01-01',
    interval = '1week')

currency_pair.head(3)
```

[Average True Range](https://docs.openbb.co/sdk/guides/intros/forex#average-true-range)

```python
weekly_atr = openbb.ta.atr(data = currency_pair, window = 4)
currency_pair = currency_pair.join(weekly_atr)

currency_pair.tail(1)
```
